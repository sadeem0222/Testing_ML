```python
import pandas as pd

data = {
    "api_calls": [5, 9, 3, 7, 2, 10, 6],
    "response_time": [200, 450, 180, 390, 150, 500, 320],
    "previous_failures": [0, 2, 0, 1, 0, 3, 1],
    "result": ["PASS", "FAIL", "PASS", "FAIL", "PASS", "FAIL", "FAIL"]
}

df = pd.DataFrame(data)

# تحويل النتيجة إلى 0 و 1
df["result"] = df["result"].map({"PASS": 0, "FAIL": 1})

df
```

```python
X = df[["api_calls", "response_time", "previous_failures"]]
y = df["result"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
new_request = [[8, 400, 2]]  # api_calls, response_time, previous_failures

prediction = model.predict(new_request)

print("FAIL" if prediction[0] == 1 else "PASS")
```
